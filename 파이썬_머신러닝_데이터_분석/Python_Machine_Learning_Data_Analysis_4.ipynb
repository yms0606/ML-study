```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/drive/MyDrive/ML study/시도별 전출입 인구수.xlsx")
```

```python
df = df.fillna(method='ffill')  #누락 값을 앞 데이터로 채움
df.head()
```

```python
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis = 1)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지',inplace=True)
df_seoul.head()
```

```python
sr_one = df_seoul.loc['경기도']
sr_one  # 서울 -> 경기도 전입 수
```

```python
plt.plot(sr_one.index, sr_one.values) # plt.plot(sr_one)
```

```python
plt.plot(sr_one.index, sr_one.values)
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
plt.figure(figsize=(14,5)) #가로 14인치, 세로 5인치
plt.plot(sr_one)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(14,5)) #가로 14인치, 세로 5인치
plt.style.use('ggplot')
plt.plot(sr_one,marker='o',markersize=10)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(14,5)) #가로 14인치, 세로 5인치
plt.style.use('ggplot')
plt.plot(sr_one,marker='o',markersize=10)
plt.xticks(rotation='vertical')
plt.ylim(50000, 800000)
plt.annotate('',xy=(20, 620000),xytext=(2,290000),xycoords='data', #끝점 , 시작점
             arrowprops=dict(arrowstyle='->',color='skyblue',lw=5))
plt.annotate('',xy=(47, 450000),xytext=(30,580000),xycoords='data',
             arrowprops=dict(arrowstyle='->',color='olive',lw=5))
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(sr_one,'o',markersize=10)
ax2.plot(sr_one,marker='o',markerfacecolor='green',markersize=10,color='olive',linewidth=2)

ax1.set_ylim(50000,800000)
ax2.set_ylim(50000,800000)

ax1.set_xticklabels(sr_one.index, rotation=75)
ax2.set_xticklabels(sr_one.index, rotation=75)

plt.show()
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML study/auto-mpg.csv', header=None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']

df.plot(kind='scatter',x='weight',y='mpg',c='coral',figsize=(10,5))
plt.show()
```

```python
df['count'] = 1
df_origin = df.groupby('origin').sum()
print(df_origin.head())
df_origin.index = ['USA','EU','JPN']

df_origin['count'].plot(kind='pie',figsize=(7,5),autopct='%1.1f%%',startangle=10,
                        colors=['chocolate','bisque','cadetblue'])
plt.show()
```

```python
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax1.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']], labels = ['USA','EU','JAPAN'])
plt.show()
```

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
sns.regplot(x='age',y='fare',data=titanic, ax=ax1)
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
sns.distplot(titanic['fare'], ax = ax1)
plt.show()
```

```python
table = titanic.pivot_table(index=['sex'],columns=['class'], aggfunc='size')

sns.heatmap(table, annot = True, fmt='d', cmap='YlGnBu',linewidth=.5,cbar=True)
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)

sns.set_style('whitegrid')

sns.stripplot(x="class",y="age",data=titanic,ax=ax1,hue='sex')
plt.show()
```
